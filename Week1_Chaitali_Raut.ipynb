```python
def safe_divide(a, b):
    # YOUR CODE HERE
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Inputs must be numeric")

    try:
        return a / b
    except ZeroDivisionError:
        return None

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None

try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"

print("1.3 passed ✓")
```

### Explanation
This function demonstrates exception handling.

- `TypeError` is raised for non-numeric inputs.
- `ZeroDivisionError` is caught and returns `None`.
- Valid numeric inputs perform normal division.

### 4.3 Eigenvalues & Eigenvectors

```python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)  # YOUR CODE HERE

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):   # YOUR CODE HERE
    v = eigenvectors[:, i]
    left = A @ v
    right = eigenvalues[i] * v

    assert np.allclose(left, right), "Eigenvector equation failed"


# Plot    # YOUR CODE HERE
fig, ax = plt.subplots()

for i in range(len(eigenvalues)):
    vec = eigenvectors[:, i] * eigenvalues[i]

    ax.quiver(
        0, 0,
        vec[0], vec[1],
        angles='xy',
        scale_units='xy',
        scale=1
    )

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

ax.set_xlabel("x")
ax.set_ylabel("y")

ax.axhline(0, color='k', lw=0.8)
ax.axvline(0, color='k', lw=0.8)

ax.set_title("Eigenvectors Scaled by Eigenvalues")

plt.tight_layout()
plt.show()

print("4.3 passed ✓")
```

### Geometric Explanation (4.3)

Eigenvectors are directions that remain unchanged after a matrix transformation.

Eigenvalues determine how much stretching or shrinking happens along those directions:

- Eigenvalue > 1 → vector stretches
- 0 < Eigenvalue < 1 → vector shrinks
- Negative eigenvalue → direction reverses

The matrix transforms space, but eigenvectors keep their original direction while only their magnitude changes.

### 4.4 SVD & Dimensionality Reduction

```python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)  # YOUR CODE HERE

X_reconstructed = U @ np.diag(S) @ Vt   # YOUR CODE HERE — reconstruct from U, S, Vt

X_approx = (
    S[0] *
    np.outer(U[:, 0], Vt[0])
)  # YOUR CODE HERE — rank-1 approximation

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"

print("4.4 passed ✓")
```

### SVD → PCA Connection Explanation(4.4)

SVD decomposes a matrix into:

- **U** → representation of samples in transformed space
- **S** → singular values showing importance of each component
- **Vt** → principal directions (feature directions)

Relation to PCA:
- PCA uses SVD internally after centering data
- Columns of `V` (rows of `Vt`) represent directions of maximum variance
- Larger singular values indicate more important components

Rank-1 approximation keeps only the strongest component and reduces dimensionality while preserving most information.

### 5.3 Error Metrics


```python
y_true = np.array([3.0, 5.0, 2.5, 7.0, 4.5, 6.0, 1.5, 8.0])
y_pred = np.array([2.8, 5.2, 2.1, 7.5, 4.0, 6.3, 2.0, 7.8])

n, p = len(y_true), 2

mae = np.mean(np.abs(y_true - y_pred))   # YOUR CODE HERE
mse = np.mean((y_true - y_pred) ** 2)    # YOUR CODE HERE
rmse = np.sqrt(mse)                      # YOUR CODE HERE

ss_res = np.sum((y_true - y_pred) ** 2)
ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)

r2 = 1 - (ss_res / ss_tot)               # YOUR CODE HERE

adj_r2 = (
    1 - (1 - r2) * (n - 1) / (n - p - 1)
)  # YOUR CODE HERE

print(f"MAE:        {mae:.4f}")
print(f"MSE:        {mse:.4f}")
print(f"RMSE:       {rmse:.4f}")
print(f"R²:         {r2:.4f}")
print(f"Adj. R²:    {adj_r2:.4f}")
```

### 5.4 Distribution Testing & Stationarity

```python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)

s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)

ks_s1 = stats.kstest(
    s1,
    'norm'
)  # YOUR CODE HERE

ks_s2 = stats.kstest(
    s2,
    'norm'
)  # YOUR CODE HERE

print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series

t = np.arange(200)

ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)   # YOUR CODE HERE

print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run

ts_diff = np.diff(ts)   # YOUR CODE HERE

adf_diff = adfuller(ts_diff)   # YOUR CODE HERE


print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
```

### Distribution Testing & Stationarity (5.4)

Kolmogorov–Smirnov Test:
- Tests whether sample data follows a reference distribution.
- p-value > 0.05 → distribution is similar.
- p-value < 0.05 → distribution differs.

ADF Test:
- Tests stationarity of a time series.

Interpretation:
- p-value < 0.05 → stationary
- p-value ≥ 0.05 → non-stationary

Differencing:
- Removes trend and helps convert non-stationary data into stationary data.

### 5.5 Model Monitoring Concepts

```python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    # YOUR CODE HERE

    epsilon = 1e-10

    breakpoints = np.linspace(
        np.min(expected),
        np.max(expected),
        bins + 1
    )

    expected_counts, _ = np.histogram(
        expected,
        bins=breakpoints
    )

    actual_counts, _ = np.histogram(
        actual,
        bins=breakpoints
    )

    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    expected_pct += epsilon
    actual_pct += epsilon

    psi = np.sum(
        (actual_pct - expected_pct) *
        np.log(actual_pct / expected_pct)
    )

    return psi

np.random.seed(1)

train_dist = np.random.normal(50, 10, 1000)
drift_dist = np.random.normal(65, 12, 1000)

psi_value = compute_psi(train_dist, drift_dist)

print(f"PSI: {psi_value:.4f}")

print(
    f"Shift severity: "
    f"{'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}"
)

# Plot both distributions

fig, ax = plt.subplots()

# YOUR CODE HERE

ax.hist(
    train_dist,
    bins=30,
    density=True,
    alpha=0.5,
    histtype='stepfilled',
    label='Training Distribution'
)

ax.hist(
    drift_dist,
    bins=30,
    density=True,
    alpha=0.5,
    histtype='stepfilled',
    label='Production Distribution'
)

ax.legend()

ax.set_xlabel("Feature Value")
ax.set_ylabel("Density")

ax.set_title('Distribution Shift Visualization')

plt.tight_layout()
plt.show()
```

### Concept Drift vs Covariate Drift & PSI Thresholds (5.5)

Concept drift:
Change in relationship between inputs and target over time.

Covariate drift:
Input feature distribution changes while target relationship stays unchanged.

PSI < 0.1:
No significant shift.

PSI 0.1–0.2:
Minor shift.

PSI > 0.2:
Major shift.

Retraining trigger example:
Retrain when PSI remains above 0.2 for multiple monitoring cycles.

## Part 6 — Probability Theory

### 6.1 Core Concepts

```python
total = 10
red, blue, green = 4, 3, 3

p_red = red / total        # YOUR CODE HERE
p_blue = blue / total      # YOUR CODE HERE
p_green = green / total    # YOUR CODE HERE


# Joint: P(red first AND blue second)

p_red_then_blue = (
    (red / total) *
    (blue / (total - 1))
)  # YOUR CODE HERE


# Conditional: P(blue second | red first)

p_blue_given_red = (
    blue / (total - 1)
)  # YOUR CODE HERE

# Independence check: are P(B|A) == P(B) ?

independent = (
    p_blue_given_red == p_blue
)  # YOUR CODE HERE

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")

print(f"P(red,blue): {p_red_then_blue:.4f}")

print(f"P(blue|red): {p_blue_given_red:.4f}")

print(f"P(blue):     {p_blue:.4f}")

print(f"Independent? {independent}")
```

### Probability Interpretation (6.1)

Sample space:
{Red, Blue, Green}

P(red):
Probability of selecting a red marble.

Joint probability:
Probability that both events occur together.

Conditional probability:
Probability of second event after first event has already occurred.

Independence:
Events are independent only if:

P(B|A) = P(B)

Here:

0.333 ≠ 0.300

### 6.3 Bayes' Theorem

```python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (
    p_free_given_spam * p_spam
    +
    p_free_given_ham * p_ham
)  # YOUR CODE HERE

# Bayes
p_spam_given_free = (
    p_free_given_spam * p_spam
) / p_free   # YOUR CODE HERE

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):

    # YOUR CODE HERE
    p_ham = 1 - prior_spam

    evidence = (
        p_word_given_spam * prior_spam
        +
        p_word_given_ham * p_ham
    )

    posterior = (
        p_word_given_spam * prior_spam
    ) / evidence

    return posterior

pred = naive_bayes_predict(0.30, 0.80, 0.05)

assert abs(pred - p_spam_given_free) < 1e-9

print(f"naive_bayes_predict: {pred:.4f}")

print("6.3 passed ✓")
```

### Bayes Term Mapping (6.3)

| Term | Formula | In this example |
|------|------|------|
| Prior | P(Spam) | Probability email is spam before seeing the word |
| Likelihood | P(word \| Spam) | Probability of seeing "free" if email is spam |
| Evidence | P(word) | Probability of seeing "free" overall |
| Posterior | P(Spam \| word) | Probability email is spam after seeing "free" |

### 6.4 Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
np.random.seed(7)

population = np.random.exponential(
    scale=1.0,
    size=100_000
)

n_samples = 5000
sample_size = 30

sample_means = np.array([
    np.mean(
        np.random.choice(
            population,
            sample_size
        )
    )
    for _ in range(n_samples)
])  # YOUR CODE HERE

pop_mean = population.mean()

pop_std = population.std()

clt_std = (
    pop_std /
    np.sqrt(sample_size)
)  # YOUR CODE HERE

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")

print(
    f"Sample means mean: "
    f"{sample_means.mean():.4f}, "
    f"std: {sample_means.std():.4f}"
)

print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()

# YOUR CODE HERE
ax.hist(
    sample_means,
    bins=35,
    density=True,
    alpha=0.7,
    label='Sample Means'
)

x = np.linspace(
    sample_means.min(),
    sample_means.max(),
    300
)

ax.plot(
    x,
    stats.norm.pdf(
        x,
        pop_mean,
        clt_std
    ),
    linewidth=3,
    label='CLT Normal Approximation'
)

ax.legend()

ax.set_title('CLT: Distribution of Sample Means')

plt.tight_layout()

plt.show()

# KS test
ks_result = stats.kstest(
    (sample_means - sample_means.mean()) /
    sample_means.std(),
    'norm'
)

print(
    f"KS test p-value: "
    f"{ks_result.pvalue:.4f} "
    f"→ Approximately normal? "
    f"{ks_result.pvalue > 0.05}"
)
```

### CLT Reflection (6.4)

This result matters because it shows that even when the original population follows a non-normal distribution, the distribution of sample means becomes closer to normal as sample size increases. In statistics and machine learning, this allows estimation and hypothesis testing using sample data instead of entire populations.

The Central Limit Theorem justifies the assumption that sampling distributions are approximately normal, which is used in many classical statistical tests such as z-tests, t-tests, and confidence intervals.
